## 2. Persamaan Matematika

##### **A. Metode Euler**
$$ \frac{dy}{dx} = f(x, y) $$
$$ \frac{dy}{dx} = \frac{y_i - y_{i+1}}{\Delta x} $$
maka solusinya: <br>
$$ y_{i+1} = y_i + f(x_{i},\space y_{i}) \Delta x $$

##### **B. Metode Heun**
Prediktor: $ y_{i+1}^{0} = y_{i} + f(x_{i},\space y_{i}) \Delta x  $ <br>
Konektor: $ y_{i+1} = y_{i} + \frac{f(x_{i},\space y_{i}) + f(x_{i+1},\space y_{i+1}^{0})}{2} \Delta x $ <br>

### **Galat:**
$$ \epsilon = Nilai\space eksak\space-\space Nilai\space numerik $$

#### 5. Source Code

```python
import numpy as np
import matplotlib.pyplot as plt
import time
import sys
from tabulate import tabulate

print("Nama : Muhammad Rosyid Suseno")
print("NIM : 1306621023")
print("Metode Euler dan Heun")
print("\n")

def fungsi(x, y):
    return (2 - x * y) * y

def solusi_eksak(x):
    return 4 / (-1 + 9 * np.exp(-2 * x) + 2 * x)

def euler(func, awal_x, awal_y, stepsize, akhir_x):
    yE = [awal_y]
    x = np.arange(awal_x, akhir_x, stepsize)
    for index, i in enumerate(x):
        if index == len(x) - 1:
            break
        yE.append(yE[index] + stepsize * func(i, yE[index]))
    return x, yE

def heun(func, awal_x, awal_y, stepsize, akhir_x):
    yH = [awal_y]
    x = np.arange(awal_x, akhir_x, stepsize)
    for index, i in enumerate(x):
        if index == len(x) - 1:
            break
        k1 = func(i, yH[index])
        k2 = func(i + stepsize, yH[index] + stepsize * k1)
        yH.append(yH[index] + (stepsize / 2) * (k1 + k2))
    return x, yH

# Nilai awal
awal_x = 0
awal_y = 0.5

# Nilai h
nilai_h = [0.05, 0.1, 0.25, 0.5, 0.7]

# Domain
akhir_x = 8

rerata_error_tabel = []
rerata_error_euler = []
rerata_error_heun = []

print('Nomor 1')

for stepsize in nilai_h:
    x_euler, y_euler = euler(fungsi, awal_x, awal_y, stepsize, akhir_x)
    x_heun, y_heun = heun(fungsi, awal_x, awal_y, stepsize, akhir_x)
    X_data = x_euler
    y_exact = solusi_eksak(X_data)
    euler_error = np.abs(np.array(y_euler) - solusi_eksak(np.array(x_euler)))
    heun_error = np.abs(np.array(y_heun) - solusi_eksak(np.array(x_heun)))
    euler_data = [[f"{x_euler[i]:.2f}", f"{y_euler[i]:.6f}"] for i in range(len(x_euler))]
    heun_data = [[f"{x_heun[i]:.2f}", f"{y_heun[i]:.6f}"] for i in range(len(x_heun))]
    euler_avg_error = np.mean(euler_error)
    heun_avg_error = np.mean(heun_error)

    euler_table = []
    heun_table = []

    for i in range(len(x_euler)):
        euler_table.append([f"{x_euler[i]:.2f}", f"{y_euler[i]:.6f}", f"{y_exact[i]:.6f}", f"{euler_error[i]:.8f}"])
    
    for i in range(len(x_heun)):
        heun_table.append([f"{x_heun[i]:.2f}", f"{y_heun[i]:.6f}", f"{y_exact[i]:.6f}", f"{heun_error[i]:.8f}"])

    rerata_error_tabel.append([stepsize, euler_avg_error, heun_avg_error])
    rerata_error_euler.append(euler_avg_error)
    rerata_error_heun.append(heun_avg_error)

    headers = ["x", "y (Euler)", "y (Eksak)", "Galat (Euler) (%)"]
    euler_table_str = tabulate(euler_table, headers, tablefmt='grid', colalign=("center", "center", "center", "center"))

    headers = ["x", "y (Heun)", "y (Eksak)", "Galat (Heun) (%)"]
    heun_table_str = tabulate(heun_table, headers, tablefmt='grid', colalign=("center", "center", "center", "center"))

    euler_start = time.process_time()
    print(f"~~~~~~~~ Solusi dengan Metode Euler (x = {stepsize}) ~~~~~~~~")
    print(euler_table_str)
    print()
    euler_end = time.process_time()
    euler_time = euler_end - euler_start

    heun_start = time.process_time()
    print(f"~~~~~~~~ Solusi dengan Metode Heun (x = {stepsize}) ~~~~~~~~")
    print(heun_table_str)
    print()
    heun_end = time.process_time()
    heun_time = heun_end - euler_start


headers = ["h", "Rata-rata Galat Euler", "Rata-rata Galat Heun"]
avg_error_table_str = tabulate(rerata_error_tabel, headers, tablefmt='grid',colalign=("center", "center", "center"))

euler_memory = sys.getsizeof(euler)
heun_memory = sys.getsizeof(heun)

print("="*92)
print("Nomor 2")
print("Tabel Rata-rata Galat:")
print(avg_error_table_str)
print("Note: Galat spesifik tiap metode terlampir dalam tabel nomor 1")
print()

plt.figure(figsize=(8, 6))
plt.plot(nilai_h, rerata_error_euler,color='b', marker='o', label='Euler')
plt.plot(nilai_h, rerata_error_heun,color='brown', marker='o', label='Heun')
plt.xlabel('Ukuran Langkah (h)')
plt.ylabel('Rata-Rata Galat')
plt.title('Grafik Rata-Rata Galat')
plt.legend()
plt.grid(True)

# Add annotations for error values
for i in range(len(nilai_h)):
    plt.annotate(f'{rerata_error_euler[i]:.6f}', (nilai_h[i], rerata_error_euler[i]), textcoords="offset points", arrowprops=dict(facecolor='blue', shrink=0.05),xytext=(0,30), ha='center')
    plt.annotate(f'{rerata_error_heun[i]:.6f}', (nilai_h[i], rerata_error_heun[i]), textcoords="offset points", xytext=(0,-15), ha='center')

plt.show()
print()
print("="*92)
print("Nomor 3")

for stepsize in nilai_h:
    x_euler, y_euler = euler(fungsi, awal_x, awal_y, stepsize, akhir_x)
    x_heun, y_heun = heun(fungsi, awal_x, awal_y, stepsize, akhir_x)
    X_data = x_euler
    y_exact = solusi_eksak(X_data)
    
    plt.plot(x_euler, y_euler, label="Metode Euler", color="red")
    plt.plot(x_heun, y_heun, label="Metode Heun", color="blue")
    plt.plot(X_data, y_exact, label="Metode Eksak", color="brown")
    plt.title(f"Hasil Numerik dan Eksak pada (x = {stepsize})")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend()
    plt.grid(True)
    plt.show()

print()
print("Analisis Program")
print("Waktu pemrosesan program dengan metode euler: ","{:.5f}".format(euler_time), "detik")
print("Waktu pemrosesan program dengan metode heun: ","{:.5f}".format(heun_time), "detik")
print("Beban memori dengan metode euler: ", euler_memory, "bytes")
print("Beban memori dengan metode heun: ", heun_memory, "bytes")
```

#### 6. Analisis & Kesimpulan

Menurut teori, metode Euler adalah metode orde pertama yang sederhana dan mudah diimplementasikan. Namun, ia cenderung memberikan estimasi yang kurang akurat terutama untuk solusi yang kompleks atau ketika langkah ukuran yang besar digunakan. Metode ini memiliki kesalahan propagasi yang lebih besar karena hanya menggunakan nilai gradien awal untuk memperkirakan nilai di langkah berikutnya.

Di sisi lain, metode Heun adalah metode orde kedua yang menggabungkan langkah-langkah perbaikan untuk meningkatkan akurasi. Metode ini menggunakan pendekatan berbasis gradien awal dan gradien rata-rata di antara dua titik dalam interval yang sama untuk memperkirakan nilai di langkah berikutnya.

Berdasarkan hasil percobaan yang saya lakukan, dapat dilihat dalam tabel nomor 2 bahwa rata-rata error yang dihasilkan pada metode Euler lebih besar dibanding metode Heun. Hal ini membuktikan bahwa hasil percobaan sejalan dengan teori.

Selain itu, jarak (h) juga mempengaruhi akurasi dan kestabilan dari kedua metode tersebut. Semakin besar jarak maka semakin besar pula errornya. Namun penjelasan dari kedua metode sebagai berikut:

Dalam metode Euler, semakin kecil nilai h (langkah yang lebih kecil), semakin akurat aproksimasi solusi yang dihasilkan. Namun, menggunakan langkah yang sangat kecil juga dapat mengakibatkan peningkatan komputasi yang signifikan, karena jumlah langkah yang harus diambil akan meningkat. Juga dapat dilihat pada waktu pemrosesan program cenderung lebih lama dibanding metode Heun. Selanjutnya, jika langkah yang terlalu besar digunakan, metode Euler dapat menghasilkan estimasi yang kurang akurat atau bahkan tidak stabil.

Metode Heun juga dipengaruhi oleh jarak (h), tetapi secara umum dapat memberikan estimasi yang lebih akurat dibandingkan metode Euler bahkan dengan langkah yang lebih besar. Ini karena metode Heun menggunakan pendekatan berbasis gradien rata-rata, yang membantu dalam mengurangi kesalahan propagasi. Dalam metode Heun, peningkatan keakuratan dengan mengurangi h masih ada, tetapi biasanya tidak sebesar dalam metode Euler.

Berdasarkan percobaan pula, beban memori yang dihasilkan cenderung sama dari kedua metode.

Kesimpulannya adalah metode Heun lebih akurat dibanding metode Euler.
